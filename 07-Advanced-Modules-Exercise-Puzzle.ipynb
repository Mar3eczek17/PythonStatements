```python
import os
import re
import glob

for folder, sub_folders,files in os.walk(file_path):
    print(f'Currently looking at {folder}')
    print('\n')
    print('The subfolders are: ')
    for sub_fold in sub_folders:
        print(f'\t Subfolder: {sub_fold}')
        
    print('\n')
    print('the files are: ')
    
    for f in files:
        print(f'\t File: {f}')
        if f.endswith('.txt'): 
            print(os.path.join(folder, f))
            with open(os.path.join(folder, f)) as fi:
                content = fi.read()
                try:
                    phone = re.search(r'\d{3}-\d{3}-\d{4}',content).group()
                except AttributeError:
                    phone = re.search(r'\d{3}-\d{3}-\d{4}',content)
        print((phone))
    print('\n')
```

```python
pattern = r'\d{3}-\d{3}-\d{4}'
```

```python
def search(file,pattern = r'\d{3}-\d{3}-\d{4}'):
    f = open(file,'r')
    text = f.read()
    
    if re.search(pattern,text):
        return re.search(pattern,text)
    else:
        return '' # pass or '' if pass will return None in results insted of ''
```

```python
results = []
```

```python
os.getcwd()+'/extracted_content_unzip_me_for_instructions/extracted_content'
```

```python
for folder,sub_folders,files in os.walk(os.getcwd()+'/extracted_content_unzip_me_for_instructions/extracted_content'):
    
    for f in files:
        full_path = folder+"/" + f
        
        results.append(search(full_path))
```
